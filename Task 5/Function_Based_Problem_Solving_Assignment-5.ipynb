```python
# Problem 1
def post_engagement(likes):
    total = 0
    for like in likes:
        total += like
    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total)
    print("Post Status:", status)

post_engagement([200,300,150,400])
```

```python
# Problem 2
def medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock(6)
```

```python
# Problem 3
def rainfall_checker(rainfall, required):
    avg = sum(rainfall)/len(rainfall)
    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_checker([70,80,65,75],60)
```

```python
# Problem 4
def duplicate_accounts(users):
    if len(users)!=len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["vishal","rahul","amit","vishal"])
```

```python
# Problem 5
def appointment(age):
    if age>=18:
        status="Eligible"
    else:
        status="Not Eligible"
    print("Patient Age:",age)
    print("Eligibility Status:",status)

appointment(21)
```

```python
# Problem 6
def premium_crops(prices):
    premium=[p for p in prices if p>2000]
    print("Premium Crops:",premium)

premium_crops([1500,2500,3200,1800])
```

```python
# Problem 7
def system_health(errors):
    if errors==0:
        status="Healthy"
    elif errors<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",errors)
    print("System Status:",status)

system_health(7)
```

```python
# Problem 8
def transaction(amount):
    if amount<=50000:
        status="Approved"
    else:
        status="Rejected"
    print("Transaction Amount:",amount)
    print("Transaction Status:",status)

transaction(60000)
```

```python
# Problem 9
def attendance_checker(att):
    percent=sum(att)/len(att)
    if percent>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    print("Attendance Percentage:",percent)
    print("Exam Eligibility:",status)

attendance_checker([80,75,90,70,85])
```

```python
# Problem 10
def electricity_bill(units):
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=100*3+(units-100)*5
    else:
        bill=100*3+100*5+(units-200)*7

    if bill<500:
        status="Low Usage"
    elif bill<=1500:
        status="Moderate Usage"
    else:
        status="High Usage"

    print("Total Bill:",bill)
    print("Usage Status:",status)

electricity_bill(250)
```
