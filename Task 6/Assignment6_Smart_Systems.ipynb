```python
# Problem 1: Smart Parking Lot Management System

def parking_system(capacity, logs):
    parked = 0
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1
    
    print("Currently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Full / Alert")
    else:
        print("Parking Status: Available")

parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
# Problem 2: Food Delivery Time Estimator

def delivery_time(distance, traffic, weather):
    
    time = distance * 5
    
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    
    print("Estimated Delivery Time:", time, "minutes")

delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def theatre_occupancy(total_seats, booked_list):
    
    booked = len(booked_list)
    
    percentage = (booked / total_seats) * 100
    
    print("Occupancy:", percentage, "%")
    
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

theatre_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification

def server_load(cpu):
    
    avg = sum(cpu)/len(cpu)
    
    print("Average CPU Load:", avg, "%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

server_load([45,60,70,85,90])
```

```python
# Problem 5: Smart Classroom Resource Usage

def classroom_usage(resources):
    
    overused = []
    
    for name, hours in resources.items():
        if hours > 8:
            overused.append(name)
    
    print("Overused Resources:", overused)
    
    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

classroom_usage({
 "Projector":6,
 "AC":9,
 "Lights":4
})
```

```python
# Problem 6: Event Registration Capacity Controller

def event_registration(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100,105)
```
