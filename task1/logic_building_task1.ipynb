```python
# 1. User Login Check
# Problem: Given username and password, check whether login is successful.

# Fixed / correct credentials stored in the system
username = "admin"
password = "1234"

# Take username and password as input from the user
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Compare the entered values with the stored credentials
if input_username == username and input_password == password:
    print("Login Successful")  # Both username and password matched
else:
    print("Invalid Credentials")  # Either username, password or both are wrong
```

```python
# 2. Pass / Fail Analyzer
# Problem: Count pass and fail students based on marks.

# List of marks for each student
marks = [45, 78, 90, 33, 60]

# Initialize counters for pass and fail students
pass_count = 0
fail_count = 0

# A student passes if marks >= 50
for mark in marks:
    if mark >= 50:
        pass_count += 1   # Increase pass counter
    else:
        fail_count += 1   # Increase fail counter

# Print a clear summary of the results
print("Student Marks:", marks)
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
# 3. Simple Data Cleaner
# Problem: Clean and standardize user names.

# Original names list with inconsistent spaces and letter cases
names = [" Alice ", "bob", " CHARLIE "]

# Empty list that will store the cleaned names
cleaned_names = []

for name in names:
    # Remove extra spaces using strip(), then convert to lowercase
    # This helps in standardizing data before further processing
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

# Show the difference between original and cleaned data
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
# 4. Message Length Analyzer
# Problem: Analyze message sizes and flag long messages.

# List of messages received from users/system
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    # Calculate the number of characters in the current message
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    # Flag messages longer than 10 characters for further review
    if length > 10:
        print("  -> This message is longer than 10 characters.")
```

```python
# 5. Error Message Detector
# Problem: Detect error messages from system logs.

# Log entries received from a system or application
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counter to keep track of how many ERROR messages are present
error_count = 0

for entry in logs:
    # If the current log entry is exactly "ERROR", increase the counter
    if entry == "ERROR":
        error_count += 1

# Print a clear summary of the log analysis
print("Logs:", logs)
print("Total 'ERROR' entries:", error_count)
```
