```python
# Problem 1: Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = 'OFF'
    def turn_on(self):
        self.status = 'ON'
        print(self.name, 'is', self.status)
light1 = SmartLight('Bedroom Light')
light1.turn_on()
```

```python
# Problem 2: Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display(self):
        print('Employee ID Card')
        print('Name:', self.name)
        print('ID:', self.emp_id)
        print('Department:', self.department)
emp1 = Employee('Rahul', 'EMP102', 'IT')
emp1.display()
```

```python
# Problem 3: Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def show(self):
        print('Contact Saved')
        print('Name:', self.name)
        print('Phone:', self.phone)
c1 = Contact('Anita', '9876543210')
c1.show()
```

```python
# Problem 4: Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display(self):
        print('Product:', self.name)
        print('Price: ₹' + str(self.price))
p1 = Product('Headphones', 2499)
p1.display()
```

```python
# Problem 5: Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def show(self):
        print('Movie:', self.name)
        print('Rating:', self.rating, '/ 5')
m1 = Movie('Inception', 4.8)
m1.show()
```

```python
# Problem 6: Delivery Address Manager
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def display(self):
        print('Delivery Details')
        print('Customer:', self.customer)
        print('Address:', self.address)
d1 = Delivery('Suman', 'Hyderabad')
d1.display()
```
